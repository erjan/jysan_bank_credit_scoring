Exclude any arrests where the charge group description is not known

```python
df_serious_2018 = in2018[in2018['Charge Group Description']!= 'Non-Criminal Detention']
```

```python
df_serious_2018 = df_serious_2018[df_serious_2018['Charge Group Description']!= 'Pre-Delinquency']
```

```python
df_serious_2018 = df_serious_2018[df_serious_2018['Charge Group Description']!= "NaN"]
```

```python
df_serious_2018['Charge Group Description'].value_counts()
```

```python
avg_age_df = pd.DataFrame(df_serious_2018.groupby(['Charge Group Description'])['Age'].mean())
#avg_age_df['z_score'] = avg_age_df.apply(zscore)
#avg_age_df['z_score_abs'] = avg_age_df['z_score'].apply(lambda score: abs(score))
#print("the largest abs value among z scores: {}" .format(avg_age_df['z_score_abs'].max()))
```
